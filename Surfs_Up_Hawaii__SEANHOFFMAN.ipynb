# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model, reflect the tables, and view all of the classes that automap found
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# # Design a query to retrieve the last 12 months of precipitation data and plot the results
# Calculate the date 1 year ago from the last data point in the database

final_date = session.query(Measurement.date).order_by(Measurement.date.desc()).all()
final_date = final_date[0][0]
print()
print('-----------------------------------------------------------------------------------------')
print(f'The last date in the Measurement table is {final_date}.')

year_prior = dt.date(2017, 8, 23) - dt.timedelta(days=365)
print(f'The date one year prior to the last date in the Measurement table is {year_prior}.')
print('-----------------------------------------------------------------------------------------')
print()

# Perform a query to retrieve the data and precipitation scores

year_prior_table = session.query(Measurement.date, Measurement.prcp).\
                        filter(Measurement.date >= year_prior).all()

# Save the query results as a Pandas DataFrame and set the index to the date column

precip = pd.DataFrame(year_prior_table, columns=['date', 'prcp'])
precip.set_index('date', inplace=True, )

# Sort the dataframe by date

precip = precip.sort_values(by=['date'])

# Use Pandas Plotting with Matplotlib to plot the data
precip.plot(figsize=(18, 6), rot=45, color='indigo', alpha=.85, legend=True)
plt.xlabel('Date Range: 8-23-2016 to 8-23-2017', labelpad=16)
plt.ylabel('Precipitation (in)', labelpad=16)
plt.title('Observed Precipitation in Hawaii', y=1.02)
plt.show()
```

```python
# Design a query to show how many stations are available in this dataset?

stacnt = session.query(Measurement).group_by(Measurement.station).count()
print()
print('-----------------------------------------------------------------------------------------')
print('There are {} stations listed in the Measurement dataset.'.format(stacnt))
print('-----------------------------------------------------------------------------------------')
print()
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.

station_obsv = session.query(Measurement.station, func.count(Measurement.tobs)).\
                group_by(Measurement.station).\
                order_by(func.count(Measurement.tobs).desc()).all()

act_station = station_obsv[0][0] 

print()
print('---------------------------------------------------------------------------------------------------')
print(f'The most active station in for measurements is {act_station}.')
print('---------------------------------------------------------------------------------------------------')
for station, count in station_obsv:
    print(f'Station ID',station,'made',count,'observations in the timeframe within the measurement table.')
print('---------------------------------------------------------------------------------------------------')
print()
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?

mintemp = session.query(func.min(Measurement.tobs)).filter(Measurement.station == act_station).all()
maxtemp = session.query(func.max(Measurement.tobs)).filter(Measurement.station == act_station).all()
meantmp = session.query(func.avg(Measurement.tobs)).filter(Measurement.station == act_station).all()

print()
print('-----------------------------------------------------------------------------------------')
print(f'The minimum temperature recorded at {act_station} was {mintemp[0][0]}.')
print(f'The maximum temperature recorded at {act_station} was {maxtemp[0][0]}.')
print(f'The average temperature recorded at {act_station} was {round(meantmp[0][0],2)}.')
print('-----------------------------------------------------------------------------------------')
print()
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

temp_stmt = session.query(Measurement.station, Measurement.date, Measurement.tobs).\
        filter(Measurement.station == act_station).\
        filter(Measurement.date > year_prior).\
        order_by(Measurement.date.desc()).statement
temperatures = pd.read_sql_query(temp_stmt, session.bind)

temperatures.hist(column='tobs', bins=12, figsize=(12, 6), color='orange')
plt.xlabel('Temperatures', labelpad=16)
plt.ylabel('Frequency of Observations', labelpad=16)
plt.title('Temperature Observations at Waihee Station in Hawaii', y=1.02)
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.

start_vac_prev = '2017-06-22'
end_vac_prev = '2017-07-03'

values = (calc_temps(start_vac_prev, end_vac_prev))

print()
print('---------------------------------------------------------------------------------------------------')
print('Temperature measurements for the selected timeframe for your vacation (6/22/2020 to 7/3/2020)')
print(f'Minimum temperature = {values[0][0]}.')
print(f'Maximum temperature = {values[0][2]}.')
print(f'Average temperature = {round(values[0][1],2)}.')
print('---------------------------------------------------------------------------------------------------')
print()
```
